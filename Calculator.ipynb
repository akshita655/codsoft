```python
from tkinter import *

root=Tk()
root.title("Simple Calculator")
root.geometry('800x200')

def calculate():
    try:
        num1=float(e1.get())
        num2=float(e2.get())
    except ValueError:
        result_label.config(textt="Please enter valid numbers.")
        return
    
    operation = op_var.get()
    
    if operation=="+":
        result=num1+num2
    elif operation=="-":
        result=num1-num2
    elif operation=="*":
        result=num1*num2
    elif operation=="/":
        if num2==0:
            result_label.config(text="Cannot divide by zero.")
            return
        result = num1/num2
    else:
        result_label.config(text="Please select an operation.")
        return
    result_label.config(text=f"Result: {result}")
l1 = Label(root,text="Number 1:")
l1.grid(row=0,column=0,padx=10,pady=5)

e1=Entry(root)
e1.grid(row=0,column=1,padx=10,pady=5)

l2 = Label(root,text="Number 2:")
l2.grid(row=1,column=0,padx=10,pady=5)

e2=Entry(root)
e2.grid(row=1,column=1,padx=10,pady=5)

op_var = StringVar()
op_var.set("+")

l3 = Label(root,text="Operation:")
l3.grid(row=2,column=0,padx=10,pady=5)

operations_frame = Frame(root)
operations_frame.grid(row=2,column=1,padx=10,pady=5)

r_add = Radiobutton(operations_frame,text="+",variable=op_var,value="+")
r_add.pack(side=LEFT)

r_sub = Radiobutton(operations_frame,text="-",variable=op_var,value="-")
r_sub.pack(side=LEFT)

r_mul = Radiobutton(operations_frame,text="*",variable=op_var,value="*")
r_mul.pack(side=LEFT)

r_div = Radiobutton(operations_frame,text="/",variable=op_var,value="/")
r_div.pack(side=LEFT)

submit_button = Button(root,text="calculate",command=calculate)
submit_button.grid(row=3,column=0,columnspan=2,pady=10)

result_label = Label(root,text="Result:")
result_label.grid(row=4,column=0,padx=2,pady=5)

root.mainloop()
```
